#### 1.Even or Odd Checker 

```python
def even_odd(num):
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
```

```python
even_odd(2)
```

```python
even_odd(3)
```

#### 3.Simple Calculator

```python
def calculator(a, b, op):
    if op == "+":
        print(a + b)
    elif op == "-":
        print(a - b)
    elif op == "*":
        print(a * b)
    elif op == "/":
        print(a / b)
```

```python
calculator(2, 10, "*")
```

#### 4.Palindrome Checker 

```python
def palindrome(word):
    if word == word[::-1]:
        print("Palindrome")
    else:
        print("Not Palindrone")
```

```python
palindrome(list)
```

```python
palindrome("madam")
```

#### 5.Count Vowels in a String 

```python
def count_vowels(text):
    count = 0
    for ch in text.lower():
        if ch in "aeiou":
            count += 1
            print(count)
```

```python
count_vowels("vinova")
```

#### 6.Fibonacci Sequence Generator 

```python
def fibonacci(n):
    a = 0
    b = 1

    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
```

```python
fibonacci(10)
```

#### 7.Prime Number Checker 

```python
def prime_number(num):
    count = 0

    for i in range(1, num + 1):
        if num % i == 0:
            count += 1

    if count == 2:
        print("Prime Number")
    else:
        print("Not a Prime Number")
```

```python
prime_number(2)
```

#### 8.List Sum and Average 

```python
def sum_average(list):
    total = sum(list)
    average = total / len(list)

    print("Sum =", total)
    print("Average =", average)
```

```python
sum_average([12,17,19,21,15])
```

#### 9.Find Grade from Percentage 

```python
def marks(m1, m2, m3):
    avg = (m1 + m2 + m3) / 3

    if avg >= 90:
        print("A")
    elif avg >= 75:
        print("B")
    elif avg >= 50:
        print("C")
    else:
        print("Fail")
```

```python
marks(75,80,85)
```

#### 10.Password Validator

```python
def validate(password):
    has_upper = False
    has_lower = False
    has_digit = False
    has_symbol = False

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        else:
            has_symbol = True

    if len(password) >= 8 and has_upper and has_lower and has_digit and has_symbol:
        print("Valid Password")
    else:
        print("Invalid Password")

```

```python
validate("Admin@234")
```

#### 11.Number of Words in a Sentence

```python
def word_count(sentence):
    words = sentence.split()
    print(len(words))
```

#### 12.Reverse Each Word 

```python
def reverse_words(sentence):
    words = sentence.split()

    for word in words:
        print(word[::-1], end=" ")
```

```python
reverse_words("this is my book")  
```

#### 13.Leap Year Checker

```python
def leap_year(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        print("Leap Year")
    else:
        print("Not Leap Year")
```

```python
leap_year(2026)
```

```python
leap_year(2028)
```

#### 14.Multiplication Table
