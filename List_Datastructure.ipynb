## Append vs Extend 

```python
l1 = ["red" , "blue"]
```

```python
l1.append(["green" , "yellow"])
```

```python
l1
```

```python
l2 = ["red" , "blue"]
```

```python
l2.extend(["green" , "yellow"])
```

```python
l2
```

## List Methods(append,extend,insert)

### 1) append() vs extend()

```python
fruits = [ 'apple' , 'banana' ]
```

```python
fruits.append(['mango' , 'grapes' , 'chikoo'])
```

```python
fruits
```

```python
fruits = [ 'apple' , 'banana' ]
```

```python
fruits.extend(['mango' , 'grapes' , 'chikoo'])
```

```python
fruits
```

### 2) insert()

```python
nums = [10,20,40,50]
```

```python
nums.insert(2,30)
```

```python
nums
```
