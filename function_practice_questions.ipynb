#### Part B: Logic Inside Functions (Level 2)

```python
def area_rect(length, width):
    area = length * width
    print(area)
```

```python
area_rect(12,10)
```

```python
def can_vote(age):
    if age >= 18:
        print("Eligible")
    else:
        print("Not Eligible")
```

```python
can_vote(17)
```

```python
def find_max(a, b):
    if a > b:
        print(a)
    else:
        print(b)
```

```python
find_max(10,25)
```

```python
def check_len(s):
    if len(s) > 10:
        print("Long")
    else:
        print("Short")
```

```python
check_len("15")
```

```python
def celsius_to_fahrenheit(c):
    f = (c * 9/5) + 32
    print(f)
```

```python
celsius_to_fahrenheit(12)
```

#### Part C: Loops & Functions (Level 3)


```python
def make_table(n):
    for i in range(1, 11):
        print(n, "x", i, "=", n * i)
```

```python
make_table(4)
```

```python
def sum_list(numbers):
    total = 0

    for num in numbers:
        total = total + num

    print(total)
```

```python
sum_list([10,20,30,40,50])
```

```python
def count_vowels(word):
    count = 0

    for ch in word.lower():
        if ch in "aeiou":
            count += 1

    print(count)
```

```python
count_vowels("anaconda")
```

```python
def countdown(x):
    while x >= 0:
        print(x)
        x -= 1
```

```python
countdown(6)
```

```python
def divisible_by_3(lst):
    for num in lst:
        if num % 3 == 0:
            print(num)
```

```python
divisible_by_3([13,15,26,21,27,30])
```

```python
def power(base, exp=2):
    print(base ** exp)
```

```python
power(5)
```

```python
power(3,4)
```

```python
def calc(a, b, operation):
    if operation == "add":
        print(a + b)

    elif operation == "sub":
        print(a - b)

    elif operation == "mul":
        print(a * b)

    elif operation == "div":
        print(a / b)
```

```python
calc(10,20,"mul")
```

```python
def validate_password(password):
    if len(password) > 8 and "@" in password:
        print("password validate")
    else:
        print("password not validate")
```

```python
validate_password("asmita@123")
```

```python
def count_even_ages(ages):
    count = 0

    for age in ages:
        if age % 2 == 0:
            count += 1

    print(count)
```

```python
count_even_ages([12,15,13,17,18])
```
