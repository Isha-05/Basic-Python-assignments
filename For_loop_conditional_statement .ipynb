# Conditional Statements Practice

### 1.Eligibility Check

```python
age = int(input('Enter Age: '))
if age >= 18:
    print('Eligible to vote')
else:
    print('Not Eligible to vote')
```

### 2.Number Classification

```python
num = int(input('Enter number: '))
if num >= 1:
    print('Positive')
elif num <= 1:
    print('Negative')
else:
    print('Zero')
```

### 3.Grade Evaluation

```python
marks = int(input('Marks: '))
if marks >= 90 and marks <= 100:
    print('A')
elif marks >= 80 and marks <= 89:
    print('B')
elif marks >= 70 and marks <= 79:
    print('C')
elif marks >= 60 and marks <= 69:
    print('D')
else:
    print('F')
```

### 4.Leap Year Checker

```python
year = int(input('Enter year : '))
if year % 400 == 0: 
     print('Leap Year')
elif year % 4 == 0:
     print('Leap Year')
elif year % 100 == 0:
     print('Not a Leap Year')
else:
     print('Not a Leap Year')
```

### 5.Max of Three Numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    print("Largest =", a)
elif b >= a and b >= c:
    print("Largest =", b)
else:
    print("Largest =", c)
```

### 6.Even or Odd

```python
integer = int(input('Enter the no.: '))
if integer % 2 == 0:
    print('even')
else:
    print('odd')
```

### 7.Login System

```python
username = str(input("Enter username: "))
password = int(input("Enter password: "))

if username == "admin" and password == "123":
    print("Login Successful")
else:
    print("Login Failed")
```

### 8.Discount Calculation

```python
price = float(input("Enter the price of the item: "))
if price > 100:
    discount = price * 0.20
elif price >= 50:
    discount = price * 0.10
else:
    discount = 0
final_price = price - discount
print("Discount =", discount)
print("Final Price =", final_price)
```

# Basic For Loop Practice Questions 

### 1.Print Numbers from 1 to 10

```python
for i in range(1, 11):
    print(i)
```

### 2.Print Each Character in a String

```python
word = input("Enter a word: ")

for ch in word:
    print(ch)
    
```

### 3.Count Up and Down

```python
for i in range(1,6):
    print(i)
for i in range(4,0,-1):
    print(i)
    
```

### 4.Print Elements of a List

```python
fruits = ["apple", "banana", "cherry","date" ]
for i in fruits:
    print(i)
    
```

### 5.Print Even Numbers Between 1 and 10

```python
for i in range(2,11,2):
    print(i)
    
```

### 6.Simple sum of a list

```python
numbers = [1, 2, 3, 4, 5]
total = 0
for num in numbers:
    total = total + num
print(total)
```

### 7.Count Down from 5 to 1

```python
for i in range(5,0,-1):
    print(i)
```

### 8.Square Each Number in a List

```python
num = [1, 2, 3, 4, 5]
for i in num:
    print(i * i)
```

### 9.Repeat a Word Multiple Times

```python
word =str(input("Enter a word: "))
N = int(input("Enter number: "))
for i in range(N):
    print(word)
```
