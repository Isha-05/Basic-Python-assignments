# While Loop Practice Questions

## Basic level:

### 1.Countdown Timer

```python
num = int(input("Enter number: "))
while num >= 0:
    print(num)
    num = num - 1
```

### 2.Simple Sum Calculator

```python
total = 0
num = int(input("Enter number (0 to stop): "))
while num != 0:
    total = total + num
    num = int(input("Enter number (0 to stop): "))
print("Sum =", total)
```

## Intermediate Level

### 3.Reverse digit

```python
num = int(input("Enter number: "))
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
print(reverse)
```

### 4.Find the Smallest Divisor

```python
num = int(input("Enter number: "))
i = 2
while i <= num:
    if num % i == 0:
        print(i)
        break
    i += 1
```

## Advanced level

### 5.Fibonacci Sequence

```python
n = int(input("Enter terms: "))
a = 0
b = 1
count = 0
while count < n:
    print(a)
    c = a + b
    a = b
    b = c
    count += 1
```

### 6.Check Palindrome Number:

```python
num = int(input("Enter number: "))
original = num
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```
